```python
data = pd.read_csv('survey lung cancer.csv')
data.head
data.info
```

```python
data.loc[data['GENDER']=='M','GENDER'] = 1
data.loc[data['GENDER']=='F','GENDER'] = 2
data['GENDER'] = data['GENDER'].astype('int')
data.head()
```

```python
data.loc[data['LUNG_CANCER']=='YES','LUNG_CANCER'] = 1
data.loc[data['LUNG_CANCER']=='NO','LUNG_CANCER'] = 2
data['LUNG_CANCER'] = data['LUNG_CANCER'].astype('int')
data.head()
#on1=no_output2=yes_Excluding gender 
```

```python
data['LUNG_CANCER'].unique()
```

```python
data.corr()
sns.heatmap(data.corr(),annot=True,fmt="0.1f")
```

```python
data.hist()
```

```python
x = data.drop('LUNG_CANCER',axis=1)
x
```

```python
y = data['LUNG_CANCER'] 
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
model = LR.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)*100
```

```python
z = LR.predict(xtest) 
z
```

```python
import gradio as gr
import numpy as np
```

```python
def LUNG_CANCER(GENDER, AGE, SMOKING, YELLOW_FINGERS, ANXIETY,
       PEER_PRESSURE, CHRONIC_DISEASE, FATIGUE , ALLERGY , WHEEZING,
       ALCOHOL_CONSUMING, COUGHING, SHORTNESS_OF_BREATH,
       SWALLOWING_DIFFICULTY, CHEST_PAIN):
 x = np.array([GENDER, AGE, SMOKING, YELLOW_FINGERS, ANXIETY,
       PEER_PRESSURE, CHRONIC_DISEASE, FATIGUE , ALLERGY , WHEEZING,
       ALCOHOL_CONSUMING, COUGHING, SHORTNESS_OF_BREATH,
       SWALLOWING_DIFFICULTY, CHEST_PAIN])
 x = np.array(x).reshape(1,-1)
# df = sc.transform(df)
 #df = sc.transform(df)
 prediction = LR.predict(x)
 #prediction = L_model.predict(x)
 prediction = float(prediction)
 return prediction
```

```python
app = gr.Interface(fn=LUNG_CANCER,
         inputs=[gr.inputs.Radio(choices=[1,2],optional=bool,label="GENDER"),
         gr.inputs.Number(label="AGE"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="SMOKING"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="YELLOW_FINGERS"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="ANXIETY"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="PEER_PRESSURE"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="CHRONIC_DISEASE"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="FATIGUE"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="ALLERGY"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="WHEEZING"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="ALCOHOL_CONSUMING"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="COUGHING"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="SHORTNESS_OF_BREATH"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="SWALLOWING_DIFFICULTY"),
         gr.inputs.Radio(choices=[1,2],optional=bool,label="CHEST_PAIN"),
         ],
     outputs= "label",
     title="Developing an ML Model for LUNG_CANCER prediction"
     )
```

```python
app.launch(show_error=True,share=True)
```
